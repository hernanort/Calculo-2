# Práctica de funciones de varias variables

Primero vamos a utilizar el comando *subs* para evaluar funciones de dos variables. El siguiente ejemplo muestra como se utiliza este comando.

## Ejemplo

Dada la función $$ f(x,y) = 4x^3y^2 - 3xy - x -4y^3 $$

Evalue las siguientes parejas ordenadas: $(5,6)$, $(-4,9)$, $(-6,-5)$ y $(3.7, 7.9)$.

### Solución 

```python
x = symbols("x")
y = symbols("y")
```

```python
f = 4*x**3*y**2-3*x*y-x-4*y**3
f
```

```python
z1=f.subs([(x,5),(y,6)])
print("La función evaluada en el punto (5, 6) es:",z1)
z2=f.subs([(x,-4),(y,9)])
print("La función evaluada en el punto (-4, 9) es:",z2)
z3=f.subs([(x,-6),(y,-5)])
print("La función evaluada en el punto (-6, -5) es:",z3)
z4=f.subs([(x,3.7),(y,7.9)])
print("La función evaluada en el punto (3.7, 7.9) es:",z4)
```

Otra manera de evaluar funciones es creaando una función en python por medio de **def**.

A continuación se ilustra como hacerlo.

```python
def f(x,y):
    z = 4*x**3*y**2-3*x*y-x-4*y**3
    return z
```

```python
f(5,6)
```

```python
f(-4,9)
```

```python
f(-6,-5)
```

```python
f(3.7,7.9)
```

```python
def evaluar(f,a,b):
    z=f.subs([(x,a),(y,b)])
    return z
```

```python
evaluar(f,5,6)
```

```python
g=(x-y+1)/(x**2+y**2)
```

```python
evaluar(g,2,1)
```

# Graficas en el espacio tridimensional

Las funciones de dos varables se pueden representar gráficamente en un sistema rectangular de tres coordenadas. Para esto utilizaremos el comando **plot3d**. A continuación se presenta un ejemplo.

## Ejemplo

Represantar graficamente la función $f(x,y) = x^2 + 4y^2$.


```python
plot3d(x**2-y**2, (x, -5, 5), (y, -5, 5))
```
